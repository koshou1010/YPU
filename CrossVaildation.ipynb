```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot
from sklearn import neighbors, datasets, svm, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, learning_curve, train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

init_notebook_mode

alldata = pd.read_csv(
    r'D:\projects\MCAD\Machine Learning\feature_delta_max.csv')
data1 = alldata.drop(['Unnamed: 0', 'S3', 'S4', 'classes'], axis=1)

alldata = pd.read_csv(r'D:\projects\MCAD\Machine Learning\feature_tmp_r_1.csv')
data2 = alldata.drop(['Unnamed: 0', 'S3', 'S4', 'classes'], axis=1)

alldata = pd.read_csv(r'D:\projects\MCAD\Machine Learning\feature_mean_3.csv')
data3 = alldata.drop(['Unnamed: 0', 'S3', 'S4'], axis=1)
data3['classes'] = data3['classes'].map(
    {'DMMP': 0, 'MS': 1, 'TIC': 2}).astype(int)

alldf = pd.concat([data1, data2, data3], axis=1)
#print(alldf)

ndarray = alldf.values
ndarray.shape
#ndarray[:,0 : -1]
Label = ndarray[:, -1]
Features = ndarray[:, 0: -1]
# print(Label.shape)
# print(Features.shape)
X = Features
Y = Label
```

```python
#Kfold=10

from sklearn.model_selection import KFold
meanlist= []

kf = KFold(n_splits=10,shuffle=True)

for train_index, test_index in kf.split(X):
    
#     print("Train:", train_index, "Validation:",test_index)
        
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    
    RSVM = svm.SVC(kernel='rbf', C = 1000)
    RSVM.fit(X_train, Y_train)
    scores = RSVM.score(X_test,Y_test)

    meanlist.append(scores)
#     print("TEST:", test_index)



print(meanlist)
print('RSVM_mean_ACC=', sum(meanlist)/len(meanlist))
arr_std = np.std(meanlist)
print('std =', arr_std)
```

```python
#Kfold=100

from sklearn.model_selection import KFold
meanlist= []

kf = KFold(n_splits=100,shuffle=True)

for train_index, test_index in kf.split(X):
    
    #print("Train:", train_index, "Validation:",test_index)
        
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    
    RSVM = svm.SVC(kernel='rbf', C = 1000)
    RSVM.fit(X_train, Y_train)
    scores = RSVM.score(X_test,Y_test)
    #print('RSVM', scores)
    meanlist.append(scores)




print(meanlist)
print('RSVM_mean_ACC=', sum(meanlist)/len(meanlist))
arr_std = np.std(meanlist)
print('std =', arr_std)
```

```python
#leave one out

from sklearn.model_selection import LeaveOneOut
meanlist= []

loo = LeaveOneOut()
loo.get_n_splits(X)
for train_index, test_index in loo.split(X):
    
#     print("Train:", train_index, "Validation:",test_index)
        
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    
    RSVM = svm.SVC(kernel='rbf', C = 1000)
    RSVM.fit(X_train, Y_train)
    scores = RSVM.score(X_test,Y_test)
#     print('RSVM', scores)
    meanlist.append(scores)


print(meanlist)
print('RSVM_mean_ACC=', sum(meanlist)/len(meanlist))
arr_std = np.std(meanlist)
print('std2 =', arr_std)
```
